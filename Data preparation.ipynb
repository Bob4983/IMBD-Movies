```python
import pandas as pd 
import numpy as np
df = pd.read_csv('tmdb-movies.csv')

def check_nan(df):
    return df.isna().sum().any()

def reset_index(df,drop):
    return df.reset_index(drop=drop)
```

```python
df.head()
```

```python
#check if there is null values or not 
check_nan(df)
```

```python
#drop nan values  
df.dropna(inplace=True)
```

```python
#check result
check_nan(df)
```

```python
df.head()
```

```python
#make new DataFrame with profits values
df_gain = pd.DataFrame()
df_gain = df.copy()
```

```python
df['release_year'].astype(int)
df.drop(['budget_adj','revenue_adj'],axis=1,inplace=True)
```

```python
df_gain.drop(['homepage','cast','director','tagline','runtime','genres','production_companies','vote_count','release_date','vote_average','budget_adj','revenue_adj','keywords','overview'],axis=1,inplace=True)
```

```python
df_gain = df_gain.groupby(['id']).max()
df_gain['release_year'] =2021 - df_gain['release_year'] 
```

```python
df_gain.rename(columns={"release_year" : "year_ago"},inplace=True)
```

```python
df_gain.groupby(['id']).max().head(20)
```

```python
#make new column to calculate total_per_year
total_per_year = ( df_gain.revenue - df_gain.budget )/df_gain.year_ago
```

```python
df_gain['total_per_year'] = total_per_year
```

```python
df_gain
```

```python
df_gain.groupby(['total_per_year']).max().tail()
```

```python
#make new DataFrame with attributes values
df_att = df[["id","imdb_id","cast","original_title","director","genres","release_year"]]
```

```python
df_att.info()
```

```python
#extract all genres one by one 
genres = df_att['genres'].str.split('|', 1).str[0].str.strip()
genres.duplicated().sum()
genres.drop_duplicates(inplace=True)
```

```python
reset_index(genres,True)
```

```python
df_att
```

<h1>What are films with the highest profits in a short period?

```python
common_films = df_gain.groupby(['total_per_year']).max().tail(10)
common_films
```

```python
common_films.set_index(['imdb_id'],inplace=True)
common_films.index
```

```python
#extract attributes of common films from df_att
entries = []
for i in common_films.index:
    entry = df_att.loc[df_att['imdb_id'] == i]
    entries.append(entry)
common_films_att = pd.concat(entries)
```

```python
common_films_att
```

```python
#merge attributes and profits 
common_films_att = common_films_att.merge(common_films)
```

<h1>What are films that got popular in a short period ? 

```python
most_popular = df[["id","popularity","imdb_id","cast","original_title","director","genres","release_year"]]
most_popular=most_popular.groupby(['popularity']).max().tail(10)
reset_index(most_popular,False)
```

```python
most_popular['release_year'] =2021 - most_popular['release_year'] 
most_popular.rename(columns={"release_year" : "year_ago"},inplace=True)
```

```python
most_popular
```

<h1>What are top-rated films?

```python
vote = df[["id","imdb_id","original_title","release_year","vote_count","vote_average"]]
vote
```

```python
top_rated = vote.groupby(['vote_average']).max().tail(10)
reset_index(top_rated,False)
```

```python
top_rated.to_csv('top_rated.csv')
most_popular.to_csv('most_popular.csv')
common_films_att.to_csv('top_10.csv')
```
